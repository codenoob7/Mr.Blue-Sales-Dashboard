```python
def normalize_date_str(val):
    try:
        # Try parsing with dayfirst=True to catch "28-02-2025" types
        parsed = pd.to_datetime(str(val), dayfirst=True, errors='coerce')
        if pd.isnull(parsed):
            return None
        return parsed.strftime('%Y/%m/%d')  # as string in uniform format
    except Exception:
        return None

present_data['Normalized Date'] = present_data['Order Date/Time'].apply(normalize_date_str)
past_data['Normalized Date'] = past_data['Order Date/Time'].apply(normalize_date_str)

def to_date_obj(val):
    try:
        return datetime.strptime(val, '%Y/%m/%d').date() if pd.notnull(val) else None
    except Exception:
        return None

present_data['Order Date/Time'] = present_data['Normalized Date'].apply(to_date_obj)
present_data.drop(columns='Normalized Date', inplace=True)

past_data['Order Date/Time'] = past_data['Normalized Date'].apply(to_date_obj)
past_data.drop(columns='Normalized Date', inplace=True)
```

```python
def get_revenue(month):
    revenue = month['Net Amount'].sum()
    return revenue
```

```python
def get_mvc(month):
    customers_list = month.groupby('Phone').agg({
        "Name": "first",
        "Address": "first",
        "Discount": "sum",
        "Net Amount": "sum",
        "Order Date/Time": "first"
    }).sort_values(by='Net Amount',ascending=False).reset_index()
    return customers_list
```

```python
present_customers = get_mvc(present_data)
present_customers = present_customers.rename(columns={'Order Date/Time': 'Date'})
present_customers
```

```python
def get_week_of_month(d):
    if pd.isnull(d): return None  # handle nulls
    first_day = d.replace(day=1)
    dom = d.day
    adjusted_dom = dom + first_day.weekday()
    return int((adjusted_dom - 1) / 7 + 1)
```

```python
from openpyxl.styles.borders import Border,Side

def apply_Border_to_DataFrame(cell):
    cell.border = Border(left=Side(style='thin',color='BFBFBF'), 
                             right=Side(style='thin',color='BFBFBF'), 
                             top=Side(style='thin',color='BFBFBF'), 
                             bottom=Side(style='thin',color='BFBFBF'))
```

```python
from openpyxl.utils.dataframe import dataframe_to_rows

def add_df_to_excel(sheet,rows,r_index,c_index,isHidden,col_space):
    for r_idx,row in enumerate(rows,r_index):
        for c_idx,value in enumerate(row,c_index):
            #for merging the cell whose titles are long, gives column space
            if c_idx != 1:
                c_idx += col_space
            cell = sheet.cell(row=r_idx, column=c_idx, value=value)
            cell.alignment = Alignment(horizontal="center",vertical="center")
            cell.font = Font('Century Gothic',bold=False,size=12)
            if isHidden:
                cell.font = Font(color="FFFFFF")
            else:
                apply_Border_to_DataFrame(cell)  
                if cell.row == r_index:
                    cell.fill = PatternFill(start_color="B2DEFF",fill_type="solid")
                    cell.font = Font('Century Gothic',bold=True,size=11,color='000000') 

        ws.row_dimensions[r_idx].height = 35
```

```python
def set_row_height(sheet,start,end,value):
    for i in range(start,end+1):
        sheet.row_dimensions[i].height = value
```

```python
import subprocess
import os

def convert_excel_to_pdf(input_path, output_dir=None):
    # If output_dir not provided, use input file's directory
    if output_dir is None:
        output_dir = os.path.dirname(input_path)

    # Command to run LibreOffice in headless mode
    command = [
        r'C:\Program Files\LibreOffice\program\soffice.exe',
        '--headless',
        '--convert-to', 'pdf',
        '--outdir', output_dir,  # Output directory
        input_path  # Input Excel file
    ]

    try:
        subprocess.run(command, check=True)
        print("Conversion successful.")
    except subprocess.CalledProcessError as e:
        print("Conversion failed:", e)
```

```python
convert_excel_to_pdf('Spreadsheets/Mr_Blue_Monthly_Reports.xlsx')
```
